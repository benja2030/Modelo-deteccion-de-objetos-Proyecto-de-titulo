```python
%matplotlib inline
from ultralytics import YOLO
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# List all the custom weights you just trained
models_to_test = [
    r"C:\Users\Benja\Desktop\Proyecto de titulo github\Machine Learning\Modelos Entrenados\yolo8n_bastones_v1.pt",
    r"C:\Users\Benja\Desktop\Proyecto de titulo github\Machine Learning\Modelos Entrenados\yolo9n_bastones_v1.pt",
    r"C:\Users\Benja\Desktop\Proyecto de titulo github\Machine Learning\Modelos Entrenados\yolo10n_bastones_v1.pt",
    r"C:\Users\Benja\Desktop\Proyecto de titulo github\Machine Learning\Modelos Entrenados\yolo11n_bastones_v1.pt",
    r"C:\Users\Benja\Desktop\Proyecto de titulo github\Machine Learning\Modelos Entrenados\yolo26n_bastones_v1.pt"
]

yaml_path = 'C:/Users/Benja/Desktop/Proyecto de titulo github/Machine Learning/White_Cane_YOLO_Ready/dataset.yaml'

results_data = []

print("Empezando benchmark...")

for path in models_to_test:
    model = YOLO(path)
    
    # Run the validation sequence
    metrics = model.val(data=yaml_path, batch=1, device=0) 
    
    # Extrae solo el nombre del archivo (ej. "yolo8n_bastones_v1") y le quita el ".pt"
    nombre_modelo = os.path.basename(path).replace('.pt', '')
    
    # Extract the core metrics
    results_data.append({
        'Model': nombre_modelo, 
        'mAP50-95': metrics.box.map,  
        'Inference Time (ms)': metrics.speed['inference'], 
        'Parameters': model.info()[0] 
    })

# Output a clean comparison table
df = pd.DataFrame(results_data)
print("\n=== RESULTADOS DEL BENCHMARK ===")
print(df.sort_values(by='mAP50-95', ascending=False))
```
